# Bipartite Graph

In Graph theory, a graph

$G(V, E)$
where,
- $V$ is set of vertices/nodes
- $E$ is set of edges

is called a bipartite graph (or bigraph) if the set of vertices $V$ can be divided into two disjoint and independent sets $X$ and $Y$ such that every edge $(x, y)$ either connects a vertex from $X$ to $Y$ or a vertex from $Y$ to $X$ and there is no edge that connects vertices of same set.

Here we will first import NetworkX module. Now, in NetworkX module the algorithms related to bipartite graphs are not imported. So, we will **explicitly** import them.

**Note:-** You can also import them using:

```python
#importing networkx module
import networkx as nx
#importing bipartite algorithms
from networkx.algorithms import bipartite
```

Now, we will create a graph.

```python
G = nx.Graph()
nodes = [1,2,3,4]
edges = [(1,2), (1,4), (1,3), (3,2), (3,4)]
G.add_nodes_from(nodes)
G.add_edges_from(edges)
nx.draw(G, with_labels = 1)
```

Now to check whether the created graph is bipartite or not, we will call a pre-defined function from ***networkx.algorithms*** library called ***is_bipartite***

### is_bipartite():-

The basic function **is_bipartite**  returns,
- **True** if graph G is bipartite
- **False** if it is not.

Syntax: **is_bipartite(G)** *or* **bipartite.is_bipartite(G)**

**Parameters:**
- G : NetworkX Graph

```python
bipartite.is_bipartite(G)
```

You can see that the created Graph **G** is not bipartite because there exists an edge (1,3) i.e. between node 1 to node 3 this goes against the definition as there should not be more than one edge connecting two vertices/nodes in a bipartite graph.

Let us take another example.

```python
B = nx.Graph()
nodes = [1,2,3,4]
edges = [(1,2), (1,4), (3,2), (3,4)]
B.add_nodes_from(nodes)
B.add_edges_from(edges)
nx.draw(B, with_labels = 1)
```

```python
bipartite.is_bipartite(B)
```

As you can see, for above graph the **is_bipartite** function returns **True** meaning the above graph is a bipartite.

**Note:-** In above graph the two disjoint set of vertices would be $X = \left\{1, 3\right\}$ and $Y = \left\{2, 4\right\}$ as there exists only a single edge for

- $(1,2)$ i.e. connecting node 1 with node 2
- $(1,4)$ i.e. connecting node 1 with node 4
- $(3,2)$ i.e. connecting node 3 with node 2
- $(3,4)$ i.e. connecting node 3 with node 4

and vice versa

### sets():-

The basic function **sets**  returns,

- A dictionary containing bipartite node sets of graph G
- **NetworkXError: Graph is not bipartite** if it the input graph is not bipartite.

Syntax: **sets(G, top_nodes=None)** *or* **bipartite.sets(G, top_nodes=None)**

**Parameters:**
- G : NetworkX Graph
- top_nodes(optional) : Container with all nodes in one bipartite node set. If not supplied it will be computed. But if more than one solution exists an exception will be raised.

```python
bipartite.sets(B)
```

You can also store this in individual variables.

```python
X, Y = bipartite.sets(B)
X = list(X)
Y = list(Y)
print("X:", X)
print("Y:", Y)
```

### degrees():-

The basic function **degrees**  returns,

- Tuple of dictionaries of the degrees of the two bipartite sets keyed by node.

Syntax: **degrees(G, nodes, weight=None)** *or* **bipartite.degrees(G, nodes, weight=None)**

**Parameters:**
- G : Bipartite NetworkX Graph
- nodes : list or container(any one of the node set of bipartite graph)
- weight(optional) : It takes string as input or None(default = None). The edge attribute that holds the numerical value used as a weight. If None, then each edge has weight 1. The degree is the sum of the edge weights adjacent to the node.
