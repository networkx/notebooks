## Matching

In Graph theory, given an undirected graph

$G(V, E)$ where,

- $V$ is set of vertices/nodes
- $E$ is set of edges

a **matching** or **independent edge set** $M$ in $G$ is a set of pairwise non-adjacent edges, containing no loops i.e. no two edges share a common vertex.

Basically, a **matching set $M$ is a set of edges without any common vertex**.

---

### Types of Matching

- **Maximal Matching:** It is a matching if every edge of graph $G$ has an non-empty intersection with atleast one edge in set of matching $M$.

                                    (Images:- Courtesy of Wikipedia)
You can see here, that the red highlighted edges fall under the category of maximal matching as every other edge in graph $G$ has a vertex associated with them.

- **Maximum Matching:** It is a matching which has largest possible number of edges which follows the definition of matching.

                                    (Images:- Courtesy of Wikipedia)
You can see here, that there are no other edges possible which follows the definition of matching other than these many number of red highlighted ones at a time.

**Note:-**
- There can be more than one maximum matching.
- Every maximum matching is maximal but it is not true other way around.


- **Perfect Matching:** It is a matching that matches all vertices of the graph. It is only possible with a graph having an even number of vertices.


- **Near-Perfect Matching:** It is a matching in which exactly one vertex of the graph is left un-matched. It is only possible with a graph having an odd number of vertices.

---

### Minimum Maximal Matching

It is a maximal matching which contains the minimum possible number of edges.

---

#### min_maximal_matching():-

The basic function **min_maximal_matching** returns,

- The minimum set of edges which follows the definition of maximal matching.

Syntax: **min_maximal_matching(G)** or **approximation.min_maximal_matching(G)**

**Parameters:**

- G : NetworkX undirected Graph

**Note:- This computes an approximate solution for the minimum maximal cardinality matching problem.**

---

Let us make a simple graph.

```python
G = nx.Graph()
nodes = [1,2,3,4,5,6]
edges = [(1,2), (1,3), (2,3), (2,4), (3,5), (4,5), (4,6)]
G.add_nodes_from(nodes)
G.add_edges_from(edges)
nx.draw(G, with_labels = 1)
```

```python
approximation.min_maximal_matching(G)
```
